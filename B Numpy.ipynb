### Math Operations and Matrix Multiplication

```python
import numpy as np

X = np.array([[1, 2], [3, 4]])
Y = np.array([[5, 6], [7, 8]])

print(X,"\n", Y)
print("\n", X * Y, "\n",  Y * X)

print("\n", X @ Y, "\n", np.dot(X, Y), "\n", np.matmul(X, Y))

print("\n", Y @ X, "\n", np.dot(Y, X), "\n", np.matmul(Y, X))
```
