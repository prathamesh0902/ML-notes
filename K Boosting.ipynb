### Difference Between Various Boosting Algorithms

| Algorithm                       | Base Learner                    | Main Idea                                                                                | Handles Sequential Errors? | Key Hyperparameters                                 | Pros                                                  | Cons                                 | Typical Use Cases                       |
| ------------------------------- | ------------------------------- | ---------------------------------------------------------------------------------------- | -------------------------- | --------------------------------------------------- | ----------------------------------------------------- | ------------------------------------ | --------------------------------------- |
| **AdaBoost**                    | Decision Stumps (1-level trees) | Increases weights of misclassified observations so later models focus on difficult cases | ✅ Yes                      | `n_estimators`, `learning_rate`                     | Simple, interpretable, effective on clean data        | Sensitive to noisy data and outliers | Binary classification, customer churn   |
| **Gradient Boosting (GBM)**     | Shallow Decision Trees          | Fits new models to the residual errors of previous models using gradient descent         | ✅ Yes                      | `n_estimators`, `learning_rate`, `max_depth`        | Flexible, works for regression and classification     | Slow training, prone to overfitting  | Sales forecasting, risk prediction      |
| **XGBoost**                     | CART Trees                      | Optimized GBM with regularization and parallel processing                                | ✅ Yes                      | `max_depth`, `eta`, `subsample`, `colsample_bytree` | Fast, regularized, handles missing values             | Many parameters to tune              | Kaggle competitions, fraud detection    |
| **LightGBM**                    | Leaf-wise Trees                 | Grows trees leaf-wise instead of level-wise for faster training                          | ✅ Yes                      | `num_leaves`, `max_depth`, `learning_rate`          | Very fast, memory efficient, handles large datasets   | Can overfit small datasets           | Click prediction, large-scale analytics |
| **CatBoost**                    | Symmetric Trees                 | Handles categorical variables using ordered boosting                                     | ✅ Yes                      | `depth`, `learning_rate`, `iterations`              | Minimal preprocessing, excellent for categorical data | Slightly slower than LightGBM        | Banking, insurance, customer analytics  |
| **Histogram Gradient Boosting** | Histogram-based Trees           | Groups continuous features into bins before splitting                                    | ✅ Yes                      | `max_iter`, `learning_rate`, `max_depth`            | Faster than traditional GBM                           | Less mature than XGBoost             | Large datasets in Scikit-learn          |
| **LogitBoost**                  | Weak Learners                   | Boosting using logistic loss optimization                                                | ✅ Yes                      | `n_estimators`, `learning_rate`                     | Good probabilistic outputs                            | Rarely used today                    | Binary classification                   |
| **BrownBoost**                  | Weak Learners                   | Reduces sensitivity to noisy labels using adaptive weighting                             | ✅ Yes                      | Time parameter                                      | More robust to noise than AdaBoost                    | Computationally expensive            | Noisy classification problems           |

### Quick Summary

| Algorithm         | Best Known For                            |
| ----------------- | ----------------------------------------- |
| **AdaBoost**      | Reweighting misclassified observations    |
| **GBM**           | Sequential residual correction            |
| **XGBoost**       | Regularization + speed + performance      |
| **LightGBM**      | Extremely fast training on large datasets |
| **CatBoost**      | Native handling of categorical features   |
| **Histogram GBM** | Efficient histogram-based boosting        |
| **LogitBoost**    | Logistic-loss boosting                    |
| **BrownBoost**    | Robustness to noisy data                  |
