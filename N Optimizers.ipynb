### Comparison of Common Optimizers in Machine Learning / Deep Learning

| Optimizer                                | Update Rule / Idea                                          | Learning Rate Adaptation | Momentum | Advantages                              | Disadvantages                              | Best Use Cases                             |
| ---------------------------------------- | ----------------------------------------------------------- | ------------------------ | -------- | --------------------------------------- | ------------------------------------------ | ------------------------------------------ |
| **Batch Gradient Descent**               | Uses entire dataset to update weights                       | ❌ No                     | ❌ No     | Stable convergence                      | Very slow on large datasets                | Small datasets                             |
| **Stochastic Gradient Descent (SGD)**    | Updates weights after each training example                 | ❌ No                     | Optional | Fast, simple, good generalization       | Noisy updates, oscillations                | Large datasets, online learning            |
| **Mini-Batch Gradient Descent**          | Updates using small batches of data                         | ❌ No                     | Optional | Efficient, GPU-friendly                 | Requires batch size tuning                 | Most deep learning tasks                   |
| **Momentum**                             | Adds fraction of previous update                            | ❌ No                     | ✅ Yes    | Faster convergence, reduces oscillation | May overshoot minima                       | CNNs, deep networks                        |
| **Nesterov Accelerated Gradient (NAG)**  | Looks ahead before computing gradient                       | ❌ No                     | ✅ Yes    | More accurate than Momentum             | Slightly more complex                      | Faster training than Momentum              |
| **AdaGrad**                              | Larger updates for rare features, smaller for frequent ones | ✅ Yes                    | ❌ No     | Great for sparse data                   | Learning rate decreases too much over time | NLP, text data                             |
| **Adadelta**                             | Improves AdaGrad using moving window of gradients           | ✅ Yes                    | ❌ No     | Prevents learning rate decay            | May converge slower                        | Sparse features                            |
| **RMSProp**                              | Uses moving average of squared gradients                    | ✅ Yes                    | ❌ No     | Faster convergence, stable              | Requires tuning                            | RNNs, non-stationary problems              |
| **Adam**                                 | Combines Momentum + RMSProp                                 | ✅ Yes                    | ✅ Yes    | Fast, robust, widely used               | Can generalize worse than SGD              | Default choice for deep learning           |
| **AdamW**                                | Adam with decoupled weight decay                            | ✅ Yes                    | ✅ Yes    | Better regularization than Adam         | Slightly more tuning                       | Transformers, modern NLP                   |
| **Nadam**                                | Adam + Nesterov Momentum                                    | ✅ Yes                    | ✅ Yes    | Faster convergence than Adam            | More computational overhead                | Deep neural networks                       |
| **AMSGrad**                              | Adam variant with corrected adaptive learning rates         | ✅ Yes                    | ✅ Yes    | Improved theoretical convergence        | Limited practical gains                    | When Adam becomes unstable                 |
| **FTRL (Follow-The-Regularized-Leader)** | Combines adaptive learning with regularization              | ✅ Yes                    | ❌ No     | Handles sparse, large-scale problems    | Complex implementation                     | CTR prediction, recommendation systems     |
| **L-BFGS**                               | Second-order optimization approximation                     | ❌ No                     | ❌ No     | Very fast convergence on small datasets | Memory intensive                           | Logistic regression, small neural networks |

### Interview-Friendly Summary

* **SGD:** Simple optimizer with strong generalization.
* **Momentum:** SGD with memory to accelerate convergence.
* **NAG:** Momentum that anticipates future position.
* **AdaGrad:** Adaptive learning rates; best for sparse features.
* **RMSProp:** Fixes AdaGrad's decaying learning rate problem.
* **Adam:** Most popular optimizer combining Momentum and RMSProp.
* **AdamW:** Adam with improved regularization; preferred for Transformers.
* **Nadam:** Adam enhanced with Nesterov momentum.
* **FTRL:** Designed for massive sparse datasets.
* **L-BFGS:** Second-order optimizer for smaller problems.


```python
# Without using any optimizer class (like SGD, Adam, etc.), gradient descent can be implemented manually in two main ways:

# Batch Gradient Descent (BGD) → Uses all training samples to compute the gradient before updating parameters.
# Stochastic Gradient Descent (SGD) → Uses one sample at a time to compute the gradient and update parameters.


# 1. Batch Gradient Descent
import numpy as np

# Example: y = 2x + 1 (simple linear regression)
X = np.array([1, 2, 3, 4])
y = np.array([3, 5, 7, 9])

m, b = 0.0, 0.0  # initial parameters
lr = 0.01        # learning rate
epochs = 1000

for _ in range(epochs):
    y_pred = m * X + b
    error = y_pred - y
    # Gradients using ALL data
    dm = (2/len(X)) * np.dot(error, X)
    db = (2/len(X)) * np.sum(error)
    # Update parameters
    m -= lr * dm
    b -= lr * db

print(f"Batch GD: m={m:.3f}, b={b:.3f}")


# 2. Stochastic Gradient Descent
import numpy as np

X = np.array([1, 2, 3, 4])
y = np.array([3, 5, 7, 9])

m, b = 0.0, 0.0
lr = 0.01
epochs = 1000

for _ in range(epochs):
    for i in range(len(X)):
        y_pred = m * X[i] + b
        error = y_pred - y[i]
        # Gradients using ONE sample
        dm = 2 * error * X[i]
        db = 2 * error
        # Update parameters
        m -= lr * dm
        b -= lr * db

print(f"SGD: m={m:.3f}, b={b:.3f}")

# If you want, I can also show Mini-Batch Gradient Descent, which is a middle ground between the two and is most common in deep learning.
# Do you want me to add that?
```

**Key Differences**

| Method   | Gradient Computation | Update Frequency     | Pros                                   | Cons                      |
|----------|----------------------|----------------------|----------------------------------------|---------------------------|
| Batch GD | All samples          | Once per epoch       | Stable convergence                     | Slow for large datasets   |
| SGD      | One sample           | Many times per epoch | Fast updates, can escape local minima  | Noisy convergence         |


```python
# Here’s a Mini-Batch Gradient Descent example — it’s the middle ground between Batch GD and SGD:

# Batch GD → all samples at once
# SGD → one sample at a time
# Mini-Batch GD → small batches (e.g., 2–64 samples) at a time


# Mini-Batch Gradient Descent Example

import numpy as np

# Example dataset: y = 2x + 1
X = np.array([1, 2, 3, 4])
y = np.array([3, 5, 7, 9])

m, b = 0.0, 0.0  # initial parameters
lr = 0.01        # learning rate
epochs = 1000
batch_size = 2   # number of samples per mini-batch

for _ in range(epochs):
    # Shuffle data each epoch for better convergence
    indices = np.arange(len(X))
    np.random.shuffle(indices)
    X = X[indices]
    y = y[indices]
    
    # Process mini-batches
    for start in range(0, len(X), batch_size):
        end = start + batch_size
        X_batch = X[start:end]
        y_batch = y[start:end]
        
        # Predictions
        y_pred = m * X_batch + b
        error = y_pred - y_batch
        
        # Gradients for the batch
        dm = (2/len(X_batch)) * np.dot(error, X_batch)
        db = (2/len(X_batch)) * np.sum(error)
        
        # Update parameters
        m -= lr * dm
        b -= lr * db

print(f"Mini-Batch GD: m={m:.3f}, b={b:.3f}")


# Why Mini-Batch is Popular

# Faster than Batch GD for large datasets.
# Less noisy than SGD, leading to smoother convergence.
# Works well with vectorized operations and GPU acceleration.


# If you want, I can also show you a visual comparison of how Batch, SGD, and Mini-Batch converge differently on the same dataset.
# Do you want me to prepare that plot?
```

```python
# Sure — here’s a visual comparison of how Batch Gradient Descent, Stochastic Gradient Descent, and Mini-Batch Gradient Descent converge on the same simple regression problem.
# We’ll plot the loss vs. iterations so you can see the stability vs. noise trade-off.
import numpy as np
import matplotlib.pyplot as plt

# Synthetic dataset: y = 2x + 1 + noise
np.random.seed(42)
X = np.linspace(0, 10, 100)
y = 2 * X * X - 15 * X + 1 + np.random.randn(100) * 2

# Learning rate and epochs
lr = 0.01
epochs = 50
batch_size = 20  # for mini-batch

# Helper: compute gradients
def compute_gradients(m, b, X_batch, y_batch):
    y_pred = m * X_batch + b
    error = y_pred - y_batch
    dm = (2 / len(X_batch)) * np.dot(error, X_batch)
    db = (2 / len(X_batch)) * np.sum(error)
    return dm, db

# Batch GD
m_b, b_b = 0, 0
loss_batch = []
for _ in range(epochs):
    dm, db = compute_gradients(m_b, b_b, X, y)
    m_b -= lr * dm
    b_b -= lr * db
    loss_batch.append(np.mean((m_b * X + b_b - y) ** 2))

# SGD
m_s, b_s = 0, 0
loss_sgd = []
for _ in range(epochs):
    for i in range(len(X)):
        dm, db = compute_gradients(m_s, b_s, X[i:i+1], y[i:i+1])
        m_s -= lr * dm
        b_s -= lr * db
    loss_sgd.append(np.mean((m_s * X + b_s - y) ** 2))

# Mini-Batch GD
m_m, b_m = 0, 0
loss_minibatch = []
for _ in range(epochs):
    indices = np.arange(len(X))
    np.random.shuffle(indices)
    X_shuffled, y_shuffled = X[indices], y[indices]
    for start in range(0, len(X), batch_size):
        end = start + batch_size
        dm, db = compute_gradients(m_m, b_m, X_shuffled[start:end], y_shuffled[start:end])
        m_m -= lr * dm
        b_m -= lr * db
    loss_minibatch.append(np.mean((m_m * X + b_m - y) ** 2))

# Plot
plt.figure(figsize=(8, 5))
plt.plot(loss_batch, label="Batch GD", linewidth=2)
plt.plot(loss_sgd, label="SGD", linewidth=2)
plt.plot(loss_minibatch, label="Mini-Batch GD", linewidth=2)
plt.xlabel("Epochs")
plt.ylabel("Mean Squared Error")
plt.title("Convergence Comparison")
plt.legend()
plt.grid(True)
plt.show()

# What you’ll see:

# Batch GD → smooth, stable convergence curve.
# SGD → noisy curve but can converge faster in early stages.
# Mini-Batch GD → less noisy than SGD, faster than Batch GD.


# If you want, I can also make a 2D contour plot showing how each method moves across the loss surface step-by-step — that’s even more intuitive visually.
# Do you want me to prepare that version too?
```
