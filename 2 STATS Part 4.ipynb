# Sampling Techniques

## Data Cleaning - Titanic data

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
df.columns
```

```python
df.drop(['alive','who','class', 'adult_male'],axis=1,inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df['age'].fillna(df['age'].mean(),inplace=True)
df.drop('deck',axis=1,inplace=True)
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df.shape
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
df.head()
```

```python
df.shape
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
# Conveninance Smapling
# age - 30 to 50   pclass==1 , fare > 30  embark_town == Queenstown
conv_sample = df.query("age > 30 and age < 50  and pclass==1 and fare >= 30 and embark_town == 'Queenstown'")
```
